### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic =pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv",sep=",")
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index(titanic.PassengerId, inplace=True)
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
male=titanic.loc[titanic.Sex=='male'].count()[0]
male
```

```python
female=titanic.loc[titanic.Sex=='female'].count()[0]
female
```

```python

plt.pie([male,female],labels=['male','female'], autopct="%.2f%%")
plt.legend()
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.scatterplot(x='Age',y='Fare',data=titanic,hue=titanic.Sex)
```
